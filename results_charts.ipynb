```python
import pickle
import pandas as pd
```

```python
with open('checkpoints/ours_with_coreness_real.pickle', 'rb') as f:
    results: dict[str, float] = pickle.load(f)

results = [
    [*key.split('-'), val] \
    for key, val in results.items()
]
results = pd.DataFrame(results, columns=['data', 'epsilon', 'strategy', 'coreness', 'score'])
results = results.astype(dict(
    epsilon=float,
    strategy=str,
    coreness=bool,
    score=float
))

results_max = results.loc[results.groupby('data')['score'].idxmax(), :].copy()
results_max = results_max.reset_index(drop=True)
```

```python
results
```

```python
results_max
```

```python
with open('./params/real.pickle', 'wb') as f:
    pickle.dump(results_max, f)
```

```python
with open('checkpoints/ours_with_corenessgg.pickle', 'rb') as f:
    results: dict[str, float] = pickle.load(f)

results = [
    [*key.split('-'), val] \
    for key, val in results.items()
]
results = pd.DataFrame(results, columns=['data_num_1', 'data', 'epsilon', 'strategy', 'coreness', 'score'])
results = results.astype(dict(
    epsilon=float,
    strategy=str,
    coreness=bool,
    score=float
))
results.data = results.data_num_1 + '-' + results.data
results = results.drop(columns=['data_num_1'])

results_max = results.loc[results.groupby('data')['score'].idxmax(), :].copy()
results_max = results_max.iloc[[0, 2, 3, 4, 5, 6, 7, 8, 9, 1], :]
results_max = results_max.reset_index(drop=True)
```

```python
with open('./params/TC1.pickle', 'wb') as f:
    pickle.dump(results_max, f)
```
